```python
class Node:
    def __init__(self, parent=None, token=None, label=None):
      self.children = []
      self.parent = parent
      self.token = token
      self.label = label
```

```python
import regex as re
with open('wsj_0001.mrg') as f:
    lines = f.readlines()
    indices = []
    for i in range(len(lines)):
        if lines[i] == '( (S \n':
            indices.append(i)
    sentences = []
    for i in range(len(indices)):
        if i == len(indices) - 1:
            sentences.append(lines[indices[i]:])
        else:
            sentences.append(lines[indices[i]:indices[i+1]])

stripped_sentences = []
for sen in sentences:
    stripped_sentence = []
    for line in sen:
        line = line.strip()
        line = re.split('([\(\)])', line)
        line = [x for x in line if (x != '' and x != ' ')]
        stripped_sentence.append(line)
    stripped_sentences.append(stripped_sentence)
flattened_sentences = []
for sen in stripped_sentences:
    fsen = [item for sublist in sen for item in sublist]
    fsen = fsen[1:len(fsen) - 1]
    flattened_sentences.append(fsen)

sentence = flattened_sentences[0]
print(sentence)
tree = Node()
current_node = tree
for i in sentence:
    if i == '(':
        new_node = Node(parent=current_node)
        current_node.children.append(new_node)
        current_node = new_node
    elif i == ')':
        current_node = current_node.parent
    else:
        i_list = i.split(" ")
        current_node.label = i_list[0]
        if len(i_list) > 1:
            current_node.token = i_list[1]


def get_spanning_tokens(subtree):
    unvisited = []
    tokens = []
    for child in subtree.children:
        unvisited.append(child)
        print("Initial Nodes (by label) in Unvisited List")
        for i in unvisited:
            print(i.label)
    while len(unvisited) != 0:
        node = unvisited.pop(0)
        print("Node (by label) Removed")
        print(node.label)
        print("Nodes Remaining (by label) in Unvisited List")
        for i in unvisited:
            print(i.label)
        if node.token is not None:
            print("My Node is a leaf node")
            tokens.append(node.token)
            print("Token associated with leaf node")
            print(node.token)
        else:
            print("My Node is not a leaf node")
            children = node.children
            children.reverse()
            for child in children:
                print("Initial Nodes (by label) in Unvisited List")
                for i in unvisited:
                    print(i.label)
                unvisited.insert(0, child)
                print("Nodes (by label) in Unvisited List after Insert")
                for i in unvisited:
                    print(i.label)
    for tok in tokens:
        print(tok)

get_spanning_tokens(tree)
#bfs print
#children = tree.children
#while children is not None:
    #for child in children:
        #print(child.label)
        
```
